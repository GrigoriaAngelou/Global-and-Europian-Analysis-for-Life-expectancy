### *Visualizing our results*
To better understand life expectancy of each of the above Southern european countries I will use a line plot.

```python
# plot of all Southern europian countries

x = [Albania, Bosnia_Herzegovina, Bulgaria, Croatia, Greece, Italy, Slovenia, Spain, Vatican, Portugal, Romania, San_Marino]
name = ('Albania', 'Bosnia_Herzegovina', 'Bulgaria', 'Croatia', 'Greece', 'Italy', 'Slovenia', 'Spain', 'Vatican', 'Portugal', 'Romania', 'San_Marino')
i=0

for place in x:
    label = 'place'
    plt.plot(place['Year'], place['Life expectancy'], label=name[i])
    plt.title('Life expectancy of ' + name[i])
    i = i + 1
    plt.legend(loc='upper left')
    plt.show()  
```

As we can see life expectancy is increasing throughout the years in every country. <br> The only countries which we can notice a big drop in specific years are Italy and Spain.
